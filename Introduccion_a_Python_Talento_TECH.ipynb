```python
iris=datasets.load_iris() #Cragamops el Dataset Iris
iris_df= pd.DataFrame(data=iris.data, columns=iris.feature_names) #convirtiendo del Array de Sklearn a DF
iris_df["target"] = iris.target
iris
```

```python
iris_df
```

```python
iris_df = iris_df.rename(columns={'target': 'species'}) #renombrar columnas
```

```python
iris_df.head()
```

```python
# @title petal length (cm) vs petal width (cm)
from matplotlib import pyplot as plt
iris_df.plot(kind='scatter', x='petal length (cm)', y='petal width (cm)', s=32, alpha=.8, color='purple')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Transformar el tipo de dtao del campo SPECIES
iris_df["species"] = iris_df["species"].astype("category")
```

```python
iris_df.info() #Ver CAMBIOSSSSSSSSSSSSSSSSSSSSS
```

```python
##Asigno Valores NULOS  a la columna ESPECIES de forma ALEATORIO

iris_amp=mf.ampute_data(iris_df, perc=0.25, random_state=5000) ## Realizo amputación de datos asignando NaN
iris_amp
```

```python
#Crear KERNEL
kds = mf.ImputationKernel(
    iris_amp,                       ##Nombre de DataFrame para entrenar Modelo
    random_state=1999               ## Vlores aleatorios para entrenar modelo
)

#Correr el algoritmo MICE durante 2 INTERACCIONES
kds.mice(2)

##RETORNA el DATASet Completo
iris_completo = kds.complete_data()
```

```python
iris_completo
```
